```python
import plotly.graph_objects as go
from model.fleet import Fleet

import random

from model.fleet import Fleet
import pandas as pd
import numpy as np
```

```python
time_now = pd.Timestamp.now(tz='US/Pacific')
end_time = time_now + pd.Timedelta(days=1)
date_range = pd.period_range(start=time_now, end=end_time, freq="15min")
dt = np.zeros(24 * 4)

for i in range(24 * 4):
    dt[i] = 15 * 60

ev_capacity = 10
initial_soc_list = []
for i in range(ev_capacity):
    initial_soc_list.append(random.random())

fleet = Fleet(ev_capacity=ev_capacity, dt=dt, initial_soc_list=initial_soc_list, date_range=date_range, slow_charge_cost=100000)
status, ev_list, price_vector = fleet.solve()
```

```python
index = np.zeros(24 * 4)

for i in range(24 * 4):
    index[i] = i * 15

fig = go.Figure()
for i in range(10):
    fig.add_trace(go.Scatter(x=index, y=ev_list[i].soc.value, name=f"soc Ev {i}"))

fig.update_xaxes(title_text ='time')
fig.update_yaxes(title_text="soc", range = [0, 1])
```

```python
index = np.zeros(24 * 4)

for i in range(24 * 4):
    index[i] = i * 15

fig = go.Figure()
for i in range(10):
    fig.add_trace(go.Scatter(x=index, y=ev_list[i].power_state.value, name=f"power Ev {i}"))

fig.update_xaxes(title_text ='time')
fig.update_yaxes(title_text="power [kW]", range = [0, 50])
```

```python
index = np.zeros(24 * 4)

for i in range(24 * 4):
    index[i] = i * 15

fig = go.Figure()
fig.add_trace(go.Scatter(x=index, y=price_vector, name=f"Variable energy price {i}"))

fig.update_xaxes(title_text ='time')
fig.update_yaxes(title_text="price [$/kWh]", range = [0, 1])
```
